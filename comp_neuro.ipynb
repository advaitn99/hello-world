```
import numpy as np
import matplotlib.pyplot as plt 
import math
```

```
C=5
R=2
tau=R*C
urest=0
trest=0
trefrac=5
v_fire=1
dt=0.1
Tsim = np.arange(0,100,dt)
U = np.zeros(len(Tsim))
I = np.zeros(len(Tsim))
U1 = np.zeros(len(Tsim))
I1 = np.zeros(len(Tsim))
U2 = np.zeros(len(Tsim))
I2 = np.zeros(len(Tsim))
U3 = np.zeros(len(Tsim))
I3 = np.zeros(len(Tsim))
```

```
for i,t in enumerate(Tsim):
  if t>=0 and t<=100:
    I[i] = 1.5
  if t>=trest:
    U[i] = U[i-1] + (-U[i-1] + I[i]*R)/tau*dt
    if U[i]>=v_fire:
      trest =t+trefrac
```

```
trest=0
for i,t in enumerate(Tsim):
  if t>=0 and t<=100:
    I1[i] = 0.75
  if t>=trest:
    U1[i] = U1[i-1] + (-U1[i-1] + I1[i]*R)/tau*dt
    if U1[i]>=v_fire:
      trest =t+trefrac
```

```
trest=0
for i,t in enumerate(Tsim):
  if t>=0 and t<=100:
    I2[i] = 5*np.sin(np.pi*(75/180)*t)
  if t>=trest:
    U2[i] = U2[i-1] + (-U2[i-1] + I2[i]*R)/tau*dt
    if U2[i]>=v_fire:
      trest =t+trefrac
```

```
trest=0
for i,t in enumerate(Tsim):
  if t>=0 and t<=100:
    I3[i] = -2*np.sin(np.pi*(30/180)*t) + 3*np.sin(np.pi*(45/180)*t) + np.cos(np.pi*(60/180)*t)
  if t>=trest:
    U3[i] = U3[i-1] + (-U3[i-1] + I3[i]*R)/tau*dt
    if U3[i]>=v_fire:
      trest =t+trefrac
```

```
Ux = np.zeros(len(Tsim))
current = np.array([0.25, 0.5, 0.75, 1, 1.5, 2, 3, 4, 5,6,7,8,9,10])
freq = np.zeros(len(current))
trest=0
for j in range(0,len(current),1):
  Ix = current[j]
  for i,t in enumerate(Tsim):
    if t>=trest:
      Ux[i] = Ux[i-1] + (-Ux[i-1] + Ix*R)/tau*dt
      if Ux[i]>=v_fire:
        trest =t+trefrac
        freq[j] += 1 
  trest=0 
  Ux = np.zeros(len(Tsim)) 
freq = freq/0.1
```

```
plt.stem(current,freq)
plt.xlabel('I')
plt.ylabel('freq')
```
